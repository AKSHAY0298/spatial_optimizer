```python
def get_tower_cost(r):
    """
    Calculates the installation cost for a tower of radius r (in km).
    Valid for r between 5 and 100.
    """
    if r < 5 or r > 100:
        raise ValueError("Radius must be strictly between 5 and 100 km.")
        
    if 5 <= r <= 20:
        return (0.00003898883009994121 * (r**3)) \
             - (0.0005848324514991181 * (r**2)) \
             + (0.000818342151675485 * r) \
             + 0.9056554967666078
             
    elif 20 < r <= 35:
        return (-0.00004679600235155791 * (r**3)) \
             + (0.004562257495590829 * (r**2)) \
             - (0.10212345679012345 * r) \
             + 1.591934156378601
             
    elif 35 < r <= 50:
        return (0.00005930629041740153 * (r**3)) \
             - (0.006578483245149912 * (r**2)) \
             + (0.2878024691358025 * r) \
             - 2.957201646090535
             
    elif 50 < r <= 100:
        return (-0.00001544973544973545 * (r**3)) \
             + (0.004634920634920635 * (r**2)) \
             - (0.27286772486772487 * r) \
             + 6.387301587301588
```

```python
# Testing a combination in our grid search
radius_1 = 20
radius_2 = 100

# Python calculates the exact linear constants before the solver starts
cost_1 = get_tower_cost(radius_1)
cost_2 = get_tower_cost(radius_2)

# Now hand cost_1 and cost_2 directly to Gurobi!
print(cost_1, cost_2)
```
